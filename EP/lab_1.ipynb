```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st 
```

```python
a = 2
b = 8
n = 43
y = 15

x = np.linspace(0, 1, 200)

pdf_1 = st.beta.pdf(x, a, b)
pdf_2 = st.beta.pdf(x, a+y, b+n-y)


plt.plot(x, pdf_1, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
plt.plot(x, pdf_2, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a+y, b+n-y))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

```python
s = 1000

posterior_vals = st.beta.rvs(a+y, b+n-y, size=s)

#Mean + SD
print(np.mean(posterior_vals))
print(np.std(posterior_vals))

#Quantile range (numpy's quantile function seems to have issues with jupyter - but percentile works)
print(np.percentile(posterior_vals, 2.5))
print(np.percentile(posterior_vals, 97.5))

#True interval
print(st.beta.interval(0.95, a+y, b+n-y))
```

Repeating the above steps using a different prior

```python
a = 8
b = 2

pdf_1 = st.beta.pdf(x, a, b)
pdf_2 = st.beta.pdf(x, a+y, b+n-y)


plt.plot(x, pdf_1, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
plt.plot(x, pdf_2, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a+y, b+n-y))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

Mixture Model (two betas)

```python
a1 = 2 
b1 = 8
a2 = 8
b2 = 2

pdf_mix = 0.75*st.beta.pdf(x, a1, b1) + 0.25*st.beta.pdf(x, a2, b2)

plt.plot(x, pdf_mix, label=r'$\alpha$ = {}, $\beta$ = {}'.format([a1, a2], [b1, b2]))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

```python
pdf_1 = st.beta.pdf(x, a1, b1)
pdf_2 = st.beta.pdf(x, a2, b2)


plt.plot(x, pdf_mix, label=r'$\alpha$ = {}, $\beta$ = {}'.format([a1, a2], [b1, b2]))
plt.plot(x, pdf_1, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a1, b1))
plt.plot(x, pdf_2, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a2, b2))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```
